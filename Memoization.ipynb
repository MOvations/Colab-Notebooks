```
# Least Recently Used cache
from functools import lru_cache
```

```
%time

# add decorator for Memozation with cache extended to 1000

@lru_cache(maxsize=1000)
def fibonacci_lru(n):
  # Check that n is a positive integer:
  if type(n) != int:
    raise ValueError("n must be a positive integer")
  if (n < 1):
    raise ValueError("n must be a positive integer")
    
  # Compute the nth term with recursion:
  if (n == 1):
    return 1
  elif (n == 2):
    return 1
  elif (n > 2):
    return (fibonacci_lru(n-1) + fibonacci_lru(n-2))
```

```
%time

# add variable to store found fibonacci values

fibonacci_cache = {}
def fibonacci_var(n):
  # Check that n is a positive integer:
  if type(n) != int:
    raise ValueError("n must be a positive integer")
  if (n < 1):
    raise ValueError("n must be a positive integer")
 
  # check to see if n is in fibonacci_cache
  if n in fibonacci_cache:
    return fibonacci_cache[n]
    
  # Compute the nth term with recursion:
  if (n == 1):
    value = 1
  elif (n == 2):
    value = 1
  elif (n > 2):
    value = (fibonacci_var(n-1) + fibonacci_var(n-2))
  
  fibonacci_cache[n] = value
  return value
```

```
%%time
for n in range(1, 10001):
  fibonacci_lru(n)

print(n, ":", fibonacci_lru(n))
```

```
%%time
for n in range(1, 10001):
  fibonacci_var(n)

print(n, ":", fibonacci_var(n))
```
